```python
import numpy as np

# Define the population distribution
population = np.random.normal(50, 10, 10000)

# Set the sample size and number of samples
sample_size = 30
num_samples = 1000

# Generate num_samples samples of size sample_size from the population
samples = np.random.choice(population, size=(num_samples, sample_size), replace=True)

# Calculate the mean of each sample
sample_means = np.mean(samples, axis=1)

# Plot the sampling distribution of the mean
import matplotlib.pyplot as plt
plt.hist(sample_means, bins=30)
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
